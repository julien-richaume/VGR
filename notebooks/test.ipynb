```python
df_titles = pd.read_csv('../data/title.basics.tsv', sep='\t')
df_names = pd.read_csv('../data/name.basics.tsv', sep='\t')
```

```python
# Nettoyage rapide (IMDb utilise '\N' pour les valeurs manquantes)
df_titles.replace('\\N', np.nan, inplace=True)
df_names.replace('\\N', np.nan, inplace=True)
```

```python
def process_imdb_data(titles_df, names_df):
    # 1. Préparation : On veut lier les noms aux films.
    # On transforme la colonne 'knownForTitles' en lignes individuelles
    # pour pouvoir faire une jointure simple.
    
    # On supprime les lignes où il n'y a pas de titres connus
    names_exploded = names_df.dropna(subset=['knownForTitles'])
    
    # On "explose" la chaîne de caractères (ex: "tt1,tt2" -> deux lignes)
    names_exploded = names_exploded.assign(tconst=names_exploded['knownForTitles'].str.split(','))
    names_exploded = names_exploded.explode('tconst')
    
    # 2. Jointure : On lie les noms des acteurs aux films
    # On ne garde que les colonnes utiles pour éviter d'alourdir le DF
    names_subset = names_exploded[['tconst', 'primaryName']]
    
    # On regroupe les noms par film (pour avoir une liste de noms par film)
    names_grouped = names_subset.groupby('tconst')['primaryName'].apply(
    lambda x: ', '.join(x.astype(str))
    ).reset_index()
    
    # 3. Fusion finale avec title_basics
    df_enriched = pd.merge(titles_df, names_grouped, on='tconst', how='left')
    
    # Remplissage des NaN pour les films sans noms associés
    df_enriched['primaryName'] = df_enriched['primaryName'].fillna('')
    
    return df_enriched
```

```python
process_imdb_data(title_basics, name_basics)
```
